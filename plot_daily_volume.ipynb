```python
start_date = pd.to_datetime(before_trade_date) 
end_date = pd.to_datetime(next_trade_date)

cbond_volume = pd.DataFrame([])
for code in cbond_list:
    price_df = w.wsi(code, "volume", start_date, end_date,usedf=True)[1]   #取出来的数据是股数，要换成手数的话得除100
    price_df.columns = [code]
    cbond_volume = pd.concat([cbond_volume,price_df],axis=1)
cbond_volume = cbond_volume.ffill()    
# cbond_rtn = cbond_volume[cbond_volume.index.time!=time(15,0,0)]
# cbond_volume = cbond_volume.loc['2022-09-09 15:00:00':] #从前一个交易日的收盘价开始截取
cbond_volume_begin_time = before_trade_date[:4]+'-' + \
                    before_trade_date[4:6]+'-' + \
                    before_trade_date[6:8]+' 15:00:00'
cbond_volume = cbond_volume.loc[cbond_volume_begin_time:]
cbond_volume
```
